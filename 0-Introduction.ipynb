## Recap

What have we learned:

- gravitational waves are emitted by accelerating massive objects, but have tiny amplitudes
- the LIGO detectors detected the first GW signal from a binary black hole merger in 2015, and have detected dozens more signals with KAGRA and Virgo since then
- GW signals from binary mergers have a distinctive oscillatory signature that can be modelled accurately (sort of)
- GW detectors use lasers to measure relative displacement in long arms of an interferometer to detect disturbances caused by passing GW signals
